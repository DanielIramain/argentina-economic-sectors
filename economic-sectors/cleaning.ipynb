```python
# Source - https://stackoverflow.com/a/37926512
# Posted by EdChum
# Retrieved 2026-04-03, License - CC BY-SA 3.0

cols = df.select_dtypes(include=[object, str]).columns
df[cols] = df[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.lower())

df
```
